```python
import tensorflow as tf
import numpy as np

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = np.array([np.ravel(x) for x in x_train]).astype(np.float32) / 225.0
x_test = np.array([np.ravel(x) for x in x_test]).astype(np.float32) / 225.0
```

```python
from cuml.model_selection import GridSearchCV
from cuml.linear_model import LogisticRegression

log_reg_model = LogisticRegression(max_iter=1000)
model = GridSearchCV(log_reg_model,
                     param_grid={"C": [0.8, 0.9, 1, 1.1, 1.2, 2],
                                #"solver": ["lbfgs", "sag", "saga"]
                                },
                     verbose=3)

model.fit(x_train, y_train)

print(model.best_params_)
print(model.best_estimator_)
print(model.best_score_)
```

```python
# From previous: {'C': 0.8}, 0.9194499969482421

from cuml.model_selection import GridSearchCV
from cuml.linear_model import LogisticRegression

log_reg_model = LogisticRegression(max_iter=1000)
model = GridSearchCV(log_reg_model,
                     param_grid={"C": np.arange(0.6, 1, 0.03),
                                #"solver": ["lbfgs", "sag", "saga"]
                                },
                     verbose=3)

model.fit(x_train, y_train)

print(model.best_params_)
print(model.best_estimator_)
print(model.best_score_)
```

```python
# From previous: {'C': 0.4}, 0.920633339881897

from cuml.model_selection import GridSearchCV
from cuml.linear_model import LogisticRegression

log_reg_model = LogisticRegression(max_iter=1000)
model = GridSearchCV(log_reg_model,
                     param_grid={"C": np.arange(0.01, 0.4, 0.03),
                                #"solver": ["lbfgs", "sag", "saga"]
                                },
                     verbose=3)

model.fit(x_train, y_train)

print(model.best_params_)
print(model.best_estimator_)
print(model.best_score_)
```

```python
#Train model with best found params
#{'C': 0.16}, 0.9216166615486145

model = LogisticRegression(c=0.16, max_iter=1000, verbose=3)
model.fit(x_train, y_train)
```

```python
#Classification report
from sklearn.metrics import classification_report

y_pred = model.predict(x_test)

print(classification_report(y_test, y_pred))
```
